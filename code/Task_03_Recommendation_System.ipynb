```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = {
    "Item": [
        "Artificial Intelligence Book",
        "Machine Learning Course",
        "Deep Learning Tutorial",
        "Python Programming Book",
        "Data Science Course"
    ],
    "Description": [
        "Learn artificial intelligence concepts and applications",
        "Comprehensive machine learning course with hands-on projects",
        "Deep learning tutorial using neural networks",
        "Python programming basics and advanced concepts",
        "Data science course covering analysis and visualization"
    ]
}

df = pd.DataFrame(data)
df
```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df["Description"])

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def recommend(item_name, top_n=2):
    index = df[df["Item"] == item_name].index[0]
    similarity_scores = list(enumerate(cosine_sim[index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    recommendations = []
    for i in similarity_scores[1:top_n+1]:
        recommendations.append(df.iloc[i[0]]["Item"])

    return recommendations
```

```python
item = "Artificial Intelligence Book"
recommended_items = recommend(item)

recommended_items
```

```python
with open("recommendations.txt", "w") as f:
    f.write(f"Recommendations for '{item}':\n")
    for rec in recommended_items:
        f.write(f"- {rec}\n")

print("Output file saved successfully.")
```
